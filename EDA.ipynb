```python
#conversion de los valores de la columna accesos por cada 100 hogares a float.
df1['Accesos por cada 100 hogares'] = df1['Accesos por cada 100 hogares'].str.replace(',', '.').astype(float)
#Filtramos para el año 2022 los trimestres 1, 2, 3, 4. Luego agrupamos por provincias y sacamos la media de los accesos por cada 100 hogares
df_filtrado = df1[(df1['Año'] == 2022) & (df1['Trimestre'].isin([1, 2, 3,4]))]
df_grupo = df_filtrado.groupby('Provincia')['Accesos por cada 100 hogares'].mean()

print(df_grupo)
```

```python
#creamos paleta de colores de searborn
palette = sns.color_palette('crest') 
#generamos un gráfico de barras utilizando el método plot.
#'bar' para indicar que se desea un gráfico de barras. Figsize establece el tamaño de la figura.
#Color determina los colores utilizados en las barras del gráfico.
df_grupo.plot(kind='bar', figsize=(8, 4),color= palette)
#Se establece las etiquetas y el titulo
plt.xlabel('Provincia')
plt.ylabel('Media de Accesos por cada 100 hogares')
plt.title('Media de Accesos por cada 100 hogares por Provincia en el año 2022 y trimestres 1, 2, 3, 4.')
plt.show()
```

```python
#transformamos la columna banda ancha fija  a enteros. Remplazamos . por ''. Transformamos astype int
df4['Banda ancha fija']= df4['Banda ancha fija'].str.replace('.', '').astype(int)
```

```python
# eliminamos los caracteres no numéricos de los valores en la columna "Dial up".
df4['Dial up'] = df4['Dial up'].str.replace('[^\d]+', '', regex=True)
df4['Dial up'] = pd.to_numeric(df4['Dial up'], errors='coerce', downcast='integer') #convertir los valores a numéricos. La opción errors='coerce' reemplaza los valores no convertibles con NaN
#y la opción downcast='integer' reduce el tipo de datos a entero si es posible.
#fillna(0) para reemplazar los valores NaN con ceros y astype(int) para convertir finalmente los valores a entero
df4['Dial up'] = df4['Dial up'].fillna(0).astype(int)
```

```python
#transformamos la columna total a enteros.Remplazamos . por ''. Transformamos astype int
df4['Total'] = df4['Total'].str.replace('.', '').astype(int)
```

```python
#Agrupamos los datos por año y calculamos la suma de cada métrica
df_grouped = df4.groupby('Año').sum()

#Creamos el gráfico de líneas, filtramos
plt.plot(df_grouped.index, df_grouped['Total'], label='Total', color= 'blue')
plt.plot(df_grouped.index, df_grouped['Banda ancha fija'], label='Banda ancha fija',color= 'lightgreen')
plt.plot(df_grouped.index, df_grouped['Dial up'], label='Dial up', color= 'orange')

#Etiquetas de los ejes y título del gráfico
plt.xlabel('Año')
plt.ylabel('Cantidad de accesos')
plt.title('Evolución de accesos por año')

#Mostramos la leyenda y ajustar el diseño
plt.legend()
plt.tight_layout()

#Mostramos el gráfico
plt.show()
```

En este grafico podemos observar que el acceso a internet fijo por banda ancha va en aumento hasta que llega a su punto optimo en el año 2021 y comienza a descender. En comparación con la Banda angosta, notamos que se ha mantenido baja durante los años.

En este gráfico evaluaremos el crecimiento o decrecimiento de los accesos por tecnologías a nivel nacional durante el tercer trimestre de los últimos tres años. Hemos seleccionado el tercer trimestre debido a que el último año evaluado es 2022 y contamos con datos hasta ese periodo.

```python
# Filtrar los datos para los trimestres 3 de cada año
df52020_filtrado = df5[(df5['Año'] == 2020) & (df5['Trimestre'] == 3)]
df52021_filtrado = df5[(df5['Año'] == 2021) & (df5['Trimestre'] == 3)]
df52022_filtrado = df5[(df5['Año'] == 2022) & (df5['Trimestre'] == 3)]

# Obtener los valores de cada tecnología para cada año
adsl_values = [df52020_filtrado['ADSL'].values[0], df52021_filtrado['ADSL'].values[0], df52022_filtrado['ADSL'].values[0]]
cablemodem_values = [df52020_filtrado['Cablemodem'].values[0], df52021_filtrado['Cablemodem'].values[0], df52022_filtrado['Cablemodem'].values[0]]
fibra_values = [df52020_filtrado['Fibra óptica'].values[0], df52021_filtrado['Fibra óptica'].values[0], df52022_filtrado['Fibra óptica'].values[0]]
wireless_values = [df52020_filtrado['Wireless'].values[0], df52021_filtrado['Wireless'].values[0], df52022_filtrado['Wireless'].values[0]]
otros_values = [df52020_filtrado['Otros'].values[0], df52021_filtrado['Otros'].values[0], df52022_filtrado['Otros'].values[0]]

# Crear el gráfico de barras vertical con cada tecnología en su propia barra
fig, ax = plt.subplots()
bar_width = 0.15
bar_positions = np.arange(len(adsl_values))

# Definir una paleta de colores con tonalidades similares
colors = sns.color_palette('mako', n_colors=5)

ax.bar(bar_positions, adsl_values, width=bar_width, label='ADSL', color=colors[0])
ax.bar(bar_positions + bar_width, cablemodem_values, width=bar_width, label='Cablemodem', color=colors[1])
ax.bar(bar_positions + 2*bar_width, fibra_values, width=bar_width, label='Fibra óptica', color=colors[2])
ax.bar(bar_positions + 3*bar_width, wireless_values, width=bar_width, label='Wireless', color=colors[3])
ax.bar(bar_positions + 4*bar_width, otros_values, width=bar_width, label='Otros', color=colors[4])

# Configurar el título y etiquetas de los ejes
ax.set_title('Acceso a Internet en Trimestre 3')
ax.set_xlabel('Año')
ax.set_ylabel('Cantidad de Accesos')

# Configurar las etiquetas del eje x y las ubicaciones de las barras
ax.set_xticks(bar_positions + 2*bar_width + bar_width/2)
ax.set_xticklabels(['2020', '2021', '2022'])

# Configurar los valores del eje y
max_value = max(adsl_values + cablemodem_values + fibra_values + wireless_values + otros_values)
ax.set_ylim(0, max_value * 1.1)  # Ajustar el rango del eje y

# Agregamos la leyenda al gráfico
ax.legend()

# Mostrar el gráfico de barras vertical con cada tecnología en su propia barra
plt.show()

```

Al analizar los datos, se puede observar un notable aumento en el uso de la tecnología de fibra óptica a lo largo del tiempo. Específicamente, en el anteúltimo trimestre del año 2022, se evidencia un crecimiento significativo en el acceso a esta tecnología, casi duplicando la cantidad de usuarios en comparación con periodos anteriores.

Este hallazgo destaca la creciente popularidad y demanda de la fibra óptica como una opción preferida para la conexión a Internet.

```python
#Filtramos los datos
df12_filtrado = df12[(df12['Año'].isin([2019, 2020, 2021, 2022])) & (df12['Trimestre'].isin([1, 2, 3, 4]))]
#Creamos una nueva columna 'Año_Periodo' combinando el año y el período
df12_filtrado['Año_Periodo'] = df12_filtrado['Año'].astype(str) + '-' + df12_filtrado['Periodo']

#Creamos una nueva columna 'Año_Periodo' combinando el año y el período
df12_filtrado['Año_Periodo'] = df12_filtrado['Año'].astype(str) + '-' + df12_filtrado['Periodo']

#Ordenamos los datos por 'Año_Periodo' en orden ascendente
df12_filtrado = df12_filtrado.sort_values('Año_Periodo')
#Creamos el gráfico de líneas con períodos ordenados
plt.figure(figsize=(8, 4))
sns.lineplot(x='Año_Periodo', y='Ingresos (miles de pesos)', data=df12_filtrado, color='blue')
plt.xlabel('Periodo')
plt.ylabel('Ingresos (miles de pesos)')
plt.title('Ingresos por Periodo')
plt.xticks(rotation=45)
#Mostramos los valores de los ingresos en el gráfico
for index, row in df12_filtrado.iterrows():
    ingresos = "${:,.0f}".format(row['Ingresos (miles de pesos)'])
    plt.text(row['Año_Periodo'], row['Ingresos (miles de pesos)'], ingresos, ha='center', va='bottom', fontsize=7)

plt.show()
```

```python
latitudes = []
longitudes = []

for provincia in df_filtrado['Provincia']:
    provincia_completa = f'{provincia}, Argentina' 

    url = f'https://api.opencagedata.com/geocode/v1/json?q={provincia_completa}&key={api_key}'
    response = requests.get(url).json()

    if response['results']:
        lat = response['results'][0]['geometry']['lat']
        lng = response['results'][0]['geometry']['lng']
        latitudes.append(lat)
        longitudes.append(lng)
    else:
        latitudes.append(None)
        longitudes.append(None)

df_filtrado['Latitud'] = latitudes
df_filtrado['Longitud'] = longitudes

```
